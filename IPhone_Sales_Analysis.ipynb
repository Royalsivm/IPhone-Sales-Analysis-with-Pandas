# IPhone Sales Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/apple_products.csv')
```

**IPhone sales analysis in India**

```python
highest_rated = data.sort_values(by=["Star Rating"],ascending=False)
highest_rated = highest_rated.head(10)
highest_rated['Product Name']
```

Let's have a look how people rated top 10 products in Flipkart

```python
iphones = highest_rated['Product Name'].value_counts()
labels = iphones.index
counts = highest_rated['Number Of Ratings']
figure = sns.barplot(x=labels, y=counts)
plt.xticks(rotation=90, va="bottom")
plt.title("Number of ratings of highest rated IPhones")
plt.show()
```

```python
ipones
```

```python
iphones = highest_rated['Product Name'].value_counts()
labels = iphones.index
counts = highest_rated['Number Of Reviews']
figure = sns.barplot(x=labels, y=counts)
plt.xlabel('Name of IPhone')
plt.xticks(rotation=90, va="bottom")
plt.title("Number of reviews of highest rated IPhones")
plt.show()
```

```python
discount_percentages = data["Discount Percentage"]
scaled_discounts = (discount_percentages - discount_percentages.min()) / (discount_percentages.max() - discount_percentages.min())
bubble_sizes = scaled_discounts * 500

figure = plt.scatter(
    x="Number Of Ratings",
    y="Sale Price",
    s=bubble_sizes,
    data=data,
)

plt.xlabel("Number of Ratings")
plt.ylabel("Sale Price")
plt.title("Relationship between sales price and number of ratings")
plt.show()
```

```python
sizes = data["Sale Price"] * 0.1
sizes.dtype
```

```python
figure = plt.scatter(
    x="Discount Percentage",
    y="Number Of Ratings",
    s=sizes.values*.045,
    data=data,
)

plt.xlabel("Discount Percentage")
plt.ylabel("Number Of Ratings")
plt.title("Relationship between Discount Percentage and Number of ratings")
plt.show()
```
